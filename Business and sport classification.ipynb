```python
#Import needed functions
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/at.csv')#Declare df as the dataset
```

```python
df
```

```python
df['business'] = df['NewsType'].apply(lambda x: 1 if x == 'business' else 0)
```

```python
df
```

```python
df.groupby('NewsType').describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.Article, df.business)
```

```python
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
x_train_count
```

```python
model = MultinomialNB() #Declare model as variable for MultinomialNB
model.fit(x_train_count, y_train)
```

```python
news = ["BEIJING: The New Development Bank plans to co-finance infrastructure projects"]
news_bus_count = cv.transform(news)
model.predict(news_bus_count)
```

```python
x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)
```

```python
from sklearn.metrics import confusion_matrix

x_test_count = cv.transform(x_test)

predictions = model.predict(x_test_count)


confusion = confusion_matrix(y_test, predictions)

tn, fp, fn, tp = confusion.ravel()

# Print the values
print("True Negatives:", tn)
print("False Positives:", fp)
print("False Negatives:", fn)
print("True Positives:", tp)
```
